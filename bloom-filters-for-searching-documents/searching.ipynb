```python
from pybloom_live import BloomFilter
import re
```

```python
posts = {
    "post1": set(re.split(r"\W+", 'The five boxing wizards jump quickly'.lower())),
    "post2": set(re.split(r"\W+", 'Pack my box with five dozen liquor jugs'.lower())),
    "post3": set(re.split(r"\W+", 'Go, lazy fat vixen; be shrewd, jump quick'.lower())),
    "post4": set(re.split(r"\W+", 'When zombies arrive, quickly fax Judge Pat'.lower())),
    "post5": set(re.split(r"\W+", 'Amazingly few discotheques provide jukeboxes'.lower())),
    "post6": set(re.split(r"\W+", 'Puzzled women bequeath jerks very exotic gifts'.lower())),
    "post7": set(re.split(r"\W+", 'The quick onyx goblin jumps over the lazy dwarf'.lower())),
    "post8": set(re.split(r"\W+", 'Brawny gods just flocked up to quiz and vex him'.lower())),
    "post9": set(re.split(r"\W+", 'Watch Jeopardy, Alex Trebek’s fun TV quiz game'.lower())),
    "post10": set(re.split(r"\W+", 'Six big devils from Japan quickly forgot how to waltz'.lower())),
    "post11": set(re.split(r"\W+", 'Five or six big jet planes zoomed quickly by the tower'.lower())),
    "post12": set(re.split(r"\W+", 'Jack amazed a few girls by dropping the antique onyx vase'.lower())),
    "post13": set(re.split(r"\W+", 'A quick movement of the enemy will jeopardize six gunboats'.lower())),
    "post14": set(re.split(r"\W+", 'Jaded zombies acted quaintly but kept driving their oxen forward'.lower())),
    "post15": set(re.split(r"\W+", 'No kidding — Lorenzo called off his trip to Mexico City just because they told him the conquistadors were extinct'.lower())),
    "post16": set(re.split(r"\W+", 'Quixotic jugglers repent; wave away fake methods and brazen mishaps'.lower())),
}
```

```python
filters = {}
for name, words in posts.items():
    filters[name] = BloomFilter(capacity=len(words), error_rate=0.1)
    for word in words:
        filters[name].add(word)
```

```python
def search(search_string):
    search_terms = re.split(r"\W+", search_string.lower())
    return [name for name, filter in filters.items() if all(term in filter for term in search_terms)]
```

```python
search("jump")
```
