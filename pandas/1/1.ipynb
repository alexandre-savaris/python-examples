```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
# Remove timezone from `Created at` and convert to datetime.
df['Created at'] = pd.to_datetime(df['Created at'].str[:19])
```

```python
# Create a new DataFrame with a `year` column extracted from the `Created at` column.
df3 = df.assign(year=df['Created at'].dt.year)
```

```python
# Group by `year` and `Lineitem name` and count the number of occurrences.
df3 = df3.groupby(['year', 'Lineitem name']).size().to_frame(name='quantity')
```

```python
# Find the product with the highest quantity sold for each year.
df3 = df3[df3.groupby(['year'])['quantity'].transform(max) == df3['quantity']]
```

```python
# Print the product with the highest quantity sold for each year.
print(df3)
```
