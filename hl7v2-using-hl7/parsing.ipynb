```python
import hl7

hl7_message_string = "MSH|^~\\&|LAB|LAB_SYS|LIS|LIS_SYS|202301011200||ORU^R01|MSG001|P|2.5\rPID|||12345||DOE^JOHN||19800101|M|||123 MAIN ST^^ANYTOWN^CA^90210\rOBX|1|NM|GLUCOSE^GLUCOSE LEVEL||100|mg/dL|60-110|N|||F"

parsed_message = hl7.parse(hl7_message_string)

msh_segment = parsed_message.segment('MSH')
print(msh_segment)

patient_id = parsed_message.segment('PID')[3]
print(patient_id)

glucose_result = parsed_message.segment('OBX')[5]
print(glucose_result)
```
